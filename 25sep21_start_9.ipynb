```python
# 1. Read the three CSV files, which contain the score of the same students in term1 for each Subject

math = pd.read_csv(r'C:\D\Python_25Sep21\M6_dataset\983_m5_datasets_v1.0\MathScoreTerm1.csv')

physics = pd.read_csv(r'C:\D\Python_25Sep21\M6_dataset\983_m5_datasets_v1.0\PhysicsScoreTerm1.csv')

dss = pd.read_csv(r'C:\D\Python_25Sep21\M6_dataset\983_m5_datasets_v1.0\DSScoreTerm1.csv')
```
