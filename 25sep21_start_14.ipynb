```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
last_7day_df = df[df.Date > df['Date'].max() - pd.to_timedelta("7day")]
last_7day_df.head()
```

```python
last_7day_df_rob =  last_7day_df[last_7day_df['Category'] == 'ROBBERY']
```

```python
last_7day_df_rob.head()
```

```python
last_7day_df_rob['Date'].value_counts()
```

```python
location_list = list(last_7day_df_rob['Location'])
```

```python
location_list
```

```python
pd.DataFrame(location_list).mean()
```

```python
import folium
```

```python
import folium
map = folium.Map(location=[37.766740, -122.418305], zoom_start=10)
fg = folium.FeatureGroup(name='rob')

for i in location_list:
    fg.add_child(folium.Marker(location=i, popup='rob', icon=folium.Icon(color='green')))
map.add_child(fg)
map.save('BasicWebMap.html')  # html file saved with icon
```

```python
# 2. For the latest 15 days of data, create a Web Map of crimes, categorized as FRAUD and GAMBLING. Change the icon to font

last_15day_df = df[df.Date > df['Date'].max() - pd.to_timedelta("15day")]
last_15day_df =  last_15day_df[(last_15day_df['Category'] == 'FRAUD') | (last_15day_df['Category'] == 'GAMBLING')]
```

```python
last_15day_df.head()
```

```python
location_list_15d = list(last_15day_df['Location'])
```

```python
for i in range(len(location_list_15d)):
    location_list_15d[i] = eval(location_list_15d[i])
```

```python
location_list_15d
```

```python
pd.DataFrame(location_list_15d).mean()
```

```python
import folium
map = folium.Map(location=[37.769393, -122.423844], zoom_start=12)
fg = folium.FeatureGroup(name='15d')

for i in location_list_15d:
    fg.add_child(folium.Marker(location=i, popup='gamb', icon=folium.Icon(color='green')))
map.add_child(fg)
map.save('BasicWebMap15d.html')  # html file saved with icon
```
