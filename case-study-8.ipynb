####### The University has data of Math, Physics and Data Structure score of sophomore students. This data is stored in different files. The University has hired a data science company to do analysis of scores and find if there is any correlation of score with age, ethnicity etc. Before the data is given to the company you have to do data wrangling.
Key issues
Ensure students identify is not revealed to the agency and only relevant data is shared

Data volume:
- In thousands, but only around 1800 records are shared in files MathScoreTerm1.csv DSScoreTerm1.csv, PhysicsScoreTerm1.csv
Business benefits:
University can get more students enrollment by improving its international ranking through personalized course/curriculum for students
Approach to Solve:
You have to use fundamentals of Numpy and Pandas covered in module 4.
1. Read the three csv files which contains the score of same students in term1 for each Subject
2. Remove the name and ethnicity column (to ensure confidentiality)
3. Fill missing score data with zero
4. Merge the three files
5. Change Sex(M/F) Column to 1/2 for further analysis
6. Store the data in new file – ScoreFinal.csv

Enhancements for code:
You can try these enhancements in code
1. Convert ethnicity to numerical value
2. Fill the missing score for a student to the average of the class

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
math = pd.read_csv(r'D:\Courses\Edureka\dataset\MathScoreTerm1.csv')
dss = pd.read_csv(r'D:\Courses\Edureka\dataset\DSScoreTerm1.csv')
physics = pd.read_csv(r'D:\Courses\Edureka\dataset\PhysicsScoreTerm1.csv')
```

```python
print(math.shape)
print(math.info())
math.head()
```

```python
print(dss.shape)
print(dss.info())
dss.head()
```

```python
print(physics.shape)
print(physics.info())
physics.head()
```

```python
#deleting Name & Ethinicity columns

math.drop(['Name', 'Ethinicity'], axis=1, inplace=True)
```

```python
dss.drop(['Name', 'Ethinicity'], axis=1, inplace=True)
physics.drop(['Name', 'Ethinicity'], axis=1, inplace=True)
```

```python
print(math.head(2))
print(dss.head(2))
print(physics.head(2))
```

```python
# checking if nan values are there
math.isna().sum()
```

```python
dss.isna().sum()
```

```python
physics.isna().sum()
```

```python
# filling nan values with zeros

math = math.fillna(0)
dss = dss.fillna(0)
physics = physics.fillna(0)
```

```python
print(math.isna().sum())
print(dss.isna().sum())
print(physics.isna().sum())
```

```python
# merging 3 files

all_subject = pd.concat([math, dss, physics]).reset_index(drop=True)
```

```python
all_subject
```

```python
#changing 'Sex' column values to numbers

all_subject['Sex'] = all_subject['Sex'].replace({'M':1, 'F':2})
```

```python
all_subject
```

```python
# storing the data in to a new csv 'ScoreFinal.csv'

all_subject.to_csv(r'D:\Courses\Edureka\dataset\ScoreFinal.csv')  
```

```python
math2 = pd.read_csv(r'D:\Courses\Edureka\dataset\MathScoreTerm1.csv')
dss2 = pd.read_csv(r'D:\Courses\Edureka\dataset\DSScoreTerm1.csv')
physics2 = pd.read_csv(r'D:\Courses\Edureka\dataset\PhysicsScoreTerm1.csv')
```

```python
# filling the navalues with class average on subject

math2['Score'] = math2['Score'].fillna(math2['Score'].mean())
```

```python
dss2['Score'] = dss2['Score'].fillna(dss2['Score'].mean())
physics2['Score'] = physics2['Score'].fillna(physics2['Score'].mean())
```

```python
all_subject_2 = pd.concat([math2, dss2, physics2]).reset_index(drop=True)
```

```python
print(all_subject_2.shape)
all_subject_2.head()
```

```python
# checking values of ethinicity column

all_subject_2.groupby('Ethinicity').agg({'Ethinicity':'count'})
```
