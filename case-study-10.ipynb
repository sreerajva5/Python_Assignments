```python
# plotting x & y with error

plt.plot(x,y, linestyle='dashed', color='g', marker='o', markersize=10, markerfacecolor='b')
plt.xlabel('x')
plt.xlim([0.5,4.5])
plt.ylabel('y')
plt.ylim([19.8, 21.2])
plt.title('Simple Data Plot')
plt.errorbar(x,y, yerr = y_error, color='g', marker=None, linestyle=None)
```

5.9: Create a dataframe from following data

'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
'female': [0, 1, 1, 0, 1],
'age': [42, 52, 36, 24, 73],
'preTestScore': [4, 24, 31, 2, 3],
'postTestScore': [25, 94, 57, 62, 70]

Draw a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age

```python
df3 = pd.DataFrame({'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
                   'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 'female': [0, 1, 1, 0, 1],
                   'age': [42, 52, 36, 24, 73], 'preTestScore': [4, 24, 31, 2, 3], 'postTestScore': [25, 94, 57, 62, 70]})
```

```python
df3
```

```python
plt.scatter(x=df3['preTestScore'], y=df3['postTestScore'], s=df3['age'])
plt.show()
```

```python
plt.scatter(data = df3, x='preTestScore', y='postTestScore', s='age')
plt.show()
```

```python
# using seaborn

sns.relplot(data = df3, x='preTestScore', y='postTestScore', size='age')
```

5.10: Draw a Scatterplot from the data in question 9 of preTestScore and postTestScore with the size = 300 and the color determined by sex

```python
#using matplotlib

plt.scatter(data = df3, x='preTestScore', y='postTestScore', c='female', s=300)
plt.show()
```

```python
# using seaborn

sns.relplot(data = df3, x='preTestScore', y='postTestScore', hue='female', size=300)
```
